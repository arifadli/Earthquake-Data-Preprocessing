#**Import Library**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#**Read Dataset**

```python
# Membaca dataset
url = "https://raw.githubusercontent.com/arifadli/DatasetRepository/main/Disertation/California/usgs_california_1968_2018.csv"
data = pd.read_csv(url)

# Misalkan 80% data pertama adalah gempa susulan, dan sisanya adalah latar belakang
threshold = int(len(data) * 0.8)
susulan = data[:threshold]
latar = data[threshold:]

# Membuat plot untuk gempa susulan dan latar belakang
plt.figure(figsize=(8, 6))
plt.scatter(susulan['longitude'], susulan['latitude'], label='Aftershocks')
plt.scatter(latar['longitude'], latar['latitude'], label='Background Seismicity')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Identifikasi Aftershocks dengan MAA')
plt.legend()
plt.show()
```
