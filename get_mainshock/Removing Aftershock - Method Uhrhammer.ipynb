```python
import pandas as pd

# Membaca dataset
url = "https://raw.githubusercontent.com/arifadli/DatasetRepository/main/Disertation/California/usgs_california_1968_2018.csv"
data = pd.read_csv(url)

# Fungsi untuk menentukan apakah gempa termasuk gempa utama berdasarkan rentang magnitudo
def is_mainshock(magnitude):
    if 4 < magnitude <= 6:
        return True
    return False

# Menambahkan kolom 'Type' yang menandai gempa utama atau bukan
data['Type'] = data['mag'].apply(lambda x: 'Mainshock' if is_mainshock(x) else 'Aftershock')

# Menghitung jumlah data dalam rentang magnitudo yang berbeda
data_ranges = [
    {'Range': '2 < M < 4', 'Jumlah Data': ((data['mag'] > 2) & (data['mag'] < 4)).sum(),
     'Gempa Utama': ((data['mag'] > 2) & (data['mag'] < 4) & (data['Type'] == 'Mainshock')).sum()},
    {'Range': '4 < M < 6', 'Jumlah Data': ((data['mag'] > 4) & (data['mag'] < 6)).sum(),
     'Gempa Utama': ((data['mag'] > 4) & (data['mag'] < 6) & (data['Type'] == 'Mainshock')).sum()},
    {'Range': '6 < M < 8', 'Jumlah Data': ((data['mag'] > 6) & (data['mag'] < 8)).sum(),
     'Gempa Utama': ((data['mag'] > 6) & (data['mag'] < 8) & (data['Type'] == 'Mainshock')).sum()}
]

# Membuat DataFrame dari hasil perhitungan
result_df = pd.DataFrame(data_ranges)
print(result_df)
```

```python
# Mengambil data untuk rentang magnitudo 4-6
mag_4_6 = data[(data['mag'] > 4) & (data['mag'] < 6)].copy()

# Fungsi untuk menentukan rentang kedalaman
def categorize_depth(depth):
    if 0 < depth < 60:
        return '0 < D < 60'
    elif 60 < depth < 300:
        return '60 < D < 300'
    else:
        return '300 < D'

# Menambahkan kolom 'Kedalaman' berdasarkan rentang kedalaman
mag_4_6.loc[:, 'Kedalaman'] = mag_4_6['depth'].apply(categorize_depth)

# Menghitung jumlah data dalam rentang kedalaman yang telah ditentukan
depth_ranges = mag_4_6.groupby('Kedalaman').size().reset_index(name='Jumlah Data')

# Menampilkan hasil
print(depth_ranges)
```
