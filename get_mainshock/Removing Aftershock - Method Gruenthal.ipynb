```python
import pandas as pd

# Membaca dataset
url = "https://raw.githubusercontent.com/arifadli/DatasetRepository/main/Disertation/California/usgs_california_1968_2018.csv"
data = pd.read_csv(url)

# Fungsi untuk menentukan apakah gempa termasuk gempa utama berdasarkan rentang magnitudo
def is_mainshock(magnitude):
    if 4 < magnitude <= 6:
        return True
    return False

# Menambahkan kolom 'Type' yang menandai gempa utama atau bukan
data['Type'] = data['mag'].apply(lambda x: 'Mainshock' if is_mainshock(x) else 'Aftershock')

# Fungsi untuk kategorisasi kedalaman
def categorize_depth(depth):
    if 0 < depth < 60:
        return '0 < D < 60'
    elif 60 <= depth < 300:
        return '60 < D < 300'
    else:
        return 'D > 300'

# Menambahkan kolom 'Rentang Kedalaman' untuk klasifikasi kedalaman
data['Rentang Kedalaman'] = data['depth'].apply(categorize_depth)

# Memfilter data berdasarkan rentang kedalaman
depth_0_60 = data[data['Rentang Kedalaman'] == '0 < D < 60']
depth_60_300 = data[data['Rentang Kedalaman'] == '60 < D < 300']
depth_above_300 = data[data['Rentang Kedalaman'] == 'D > 300']

# Menghitung jumlah gempa utama (berdasarkan rentang magnitudo) dalam setiap rentang kedalaman
mainshock_depth_0_60 = depth_0_60[depth_0_60['Type'] == 'Mainshock']['Type'].count()
mainshock_depth_60_300 = depth_60_300[depth_60_300['Type'] == 'Mainshock']['Type'].count()
mainshock_depth_above_300 = depth_above_300[depth_above_300['Type'] == 'Mainshock']['Type'].count()

# Menampilkan jumlah gempa utama dalam setiap rentang kedalaman
print("Jumlah Gempa Utama dalam Rentang Kedalaman:")
print("0 < D < 60:", mainshock_depth_0_60)
print("60 < D < 300:", mainshock_depth_60_300)
print("D > 300:", mainshock_depth_above_300)
```

```python
import pandas as pd

# Membaca dataset
url = "https://raw.githubusercontent.com/arifadli/DatasetRepository/main/Disertation/California/usgs_california_1968_2018.csv"
data = pd.read_csv(url)

# Fungsi untuk menentukan apakah gempa termasuk gempa utama berdasarkan rentang magnitudo
def is_mainshock(magnitude):
    if 4 < magnitude <= 6:
        return True
    return False

# Menambahkan kolom 'Type' yang menandai gempa utama atau bukan
data['Type'] = data['mag'].apply(lambda x: 'Mainshock' if is_mainshock(x) else 'Aftershock')

# Fungsi untuk kategorisasi magnitudo
def categorize_magnitude(magnitude):
    if 2 < magnitude <= 4:
        return '2 < M < 4'
    elif 4 < magnitude <= 6:
        return '4 < M < 6'
    elif 6 < magnitude <= 8:
        return '6 < M < 8'
    else:
        return 'M > 8'

# Menambahkan kolom 'Rentang Magnitudo' untuk klasifikasi magnitudo
data['Rentang Magnitudo'] = data['mag'].apply(categorize_magnitude)

# Menghitung jumlah gempa utama (berdasarkan rentang magnitudo) dalam setiap rentang kedalaman
mainshock_magnitude_2_4 = data.loc[(data['Rentang Magnitudo'] == '2 < M < 4') & (data['Type'] == 'Mainshock'), 'Type'].count()
mainshock_magnitude_4_6 = data.loc[(data['Rentang Magnitudo'] == '4 < M < 6') & (data['Type'] == 'Mainshock'), 'Type'].count()
mainshock_magnitude_6_8 = data.loc[(data['Rentang Magnitudo'] == '6 < M < 8') & (data['Type'] == 'Mainshock'), 'Type'].count()
mainshock_magnitude_above_8 = data.loc[(data['Rentang Magnitudo'] == 'M > 8') & (data['Type'] == 'Mainshock'), 'Type'].count()

# Menampilkan jumlah gempa utama dalam setiap rentang magnitudo
print("Jumlah Gempa Utama dalam Rentang Magnitudo:")
print("2 < M < 4:", mainshock_magnitude_2_4)
print("4 < M < 6:", mainshock_magnitude_4_6)
print("6 < M < 8:", mainshock_magnitude_6_8)
print("M > 8:", mainshock_magnitude_above_8)
```
