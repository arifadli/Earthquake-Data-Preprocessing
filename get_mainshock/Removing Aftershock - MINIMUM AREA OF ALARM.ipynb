#**Read Dataset**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca dataset
url = "https://raw.githubusercontent.com/arifadli/DatasetRepository/main/Disertation/California/usgs_california_1968_2018.csv"
data = pd.read_csv(url)

threshold = int(len(data) * 0.2)
foreshock = data.iloc[0]
aftershock = data[1:threshold]
mainshock = data[threshold:]

# Membuat plot untuk gempa bumi
plt.figure(figsize=(12, 10))

plt.scatter(foreshock['longitude'], foreshock['latitude'], label='Foreshock', color='orange', marker='o', s=50)
plt.scatter(aftershock['longitude'], aftershock['latitude'], label='Aftershock', color='blue', marker='o', alpha=0.5)
plt.scatter(mainshock['longitude'], mainshock['latitude'], label='Mainshock', color='green', marker='o', alpha=0.5)

plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Identifikasi Gempa Bumi dengan MAA')
plt.legend()
plt.grid(True)
plt.show()
```
